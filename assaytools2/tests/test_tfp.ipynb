```python
import sys
sys.path.append('..')
import warnings
if not sys.warnoptions:
    warnings.simplefilter("ignore")
import numpy as np
np.warnings.filterwarnings('ignore')
import matplotlib
from matplotlib import pyplot as plt
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import time
from titration import *
from bindingmodels import *
import tensorflow as tf
tf.enable_eager_execution()
import tensorflow_probability as tfp
tfd = tfp.distributions
```

```python
print([tf.zeros(10), tf.ones(10)])
```

```python
data = np.random.normal(np.zeros(10), np.ones(10))
data = tf.constant(data, dtype=tf.float32)


def joint_log_prob(loc, scale, data=data):
    rv = tfd.Normal(loc=loc, scale=scale)
    return rv.log_prob(data)

chain = tfp.mcmc.sample_chain(
        num_results=int(10e3),
        num_burnin_steps=int(1e3),
        current_state=[tf.zeros(10), tf.ones(10)],
        kernel=tfp.mcmc.TransformedTransitionKernel(
            inner_kernel=tfp.mcmc.HamiltonianMonteCarlo(
                target_log_prob_fn=joint_log_prob,
                num_leapfrog_steps=2,
                step_size=tf.Variable(1.),
                step_size_update_fn=tfp.mcmc.make_simple_step_size_update_policy()),
            bijector=[tfp.bijectors.Exp(), tfp.bijectors.Exp()]))



```
